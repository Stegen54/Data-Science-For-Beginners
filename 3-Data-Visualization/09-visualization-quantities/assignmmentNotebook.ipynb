# Import Libraries and Load Data
Import pandas, matplotlib, seaborn and load the birds.csv dataset. Filter for Snow Geese and related waterfowl species.

```python
# Import Libraries and Load Data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
birds = pd.read_csv('../../data/birds.csv')

# Filter for Snow Geese and related waterfowl species
waterfowl_species = birds[birds['Category'] == 'Ducks/Geese/Waterfowl']
snow_geese = waterfowl_species[waterfowl_species['Name'] == "Snow goose"]

# Display the first few rows of the filtered dataset
waterfowl_species.head()
```

# Snow Goose Size Comparisons
Create bar charts comparing Snow Goose dimensions (length, wingspan, mass) to other geese species to visualize relative size differences.

```python
# Snow Goose Size Comparisons

# Create bar charts comparing Snow Goose dimensions (length, wingspan, mass) to other geese species to visualize relative size differences.

# Filter for geese species
geese_species = waterfowl_species[waterfowl_species['Name'].str.contains('goose')]

# Set plot style
sns.set(style="whitegrid")

# Plot comparison of lengths
plt.figure(figsize=(12, 6))
sns.barplot(x='Name', y='MaxLength', data=geese_species, palette='viridis')
plt.title('Comparison of Maximum Lengths of Geese Species')
plt.ylabel('Max Length (cm)')
plt.xlabel('Geese Species')
plt.xticks(rotation=45)
plt.show()

# Plot comparison of wingspans
plt.figure(figsize=(12, 6))
sns.barplot(x='Name', y='MaxWingspan', data=geese_species, palette='viridis')
plt.title('Comparison of Maximum Wingspans of Geese Species')
plt.ylabel('Max Wingspan (cm)')
plt.xlabel('Geese Species')
plt.xticks(rotation=45)
plt.show()

# Plot comparison of body masses
plt.figure(figsize=(12, 6))
sns.barplot(x='Name', y='MaxBodyMass', data=geese_species, palette='viridis')
plt.title('Comparison of Maximum Body Masses of Geese Species')
plt.ylabel('Max Body Mass (g)')
plt.xlabel('Geese Species')
plt.xticks(rotation=45)
plt.show()
```

# Wingspan vs Body Mass Analysis
Generate scatter plots looking at the relationship between wingspan and body mass across waterfowl species, highlighting where Snow Geese fall.

```python
# Wingspan vs Body Mass Analysis

# Scatter plot of wingspan vs body mass for waterfowl species
plt.figure(figsize=(12, 6))
sns.scatterplot(x='MaxBodyMass', y='MaxWingspan', data=waterfowl_species, hue='Name', palette='viridis', legend=False)
plt.title('Wingspan vs Body Mass for Waterfowl Species')
plt.xlabel('Max Body Mass (g)')
plt.ylabel('Max Wingspan (cm)')

# Highlight Snow Geese
plt.scatter(snow_geese['MaxBodyMass'], snow_geese['MaxWingspan'], color='red', s=100, label='Snow Goose')
plt.legend()
plt.show()

# Scatter plot of wingspan vs body mass for waterfowl species with log scale
plt.figure(figsize=(12, 6))
sns.scatterplot(x='MaxBodyMass', y='MaxWingspan', data=waterfowl_species, hue='Name', palette='viridis', legend=False)
plt.xscale('log')
plt.yscale('log')
plt.title('Wingspan vs Body Mass for Waterfowl Species (Log Scale)')
plt.xlabel('Max Body Mass (g)')
plt.ylabel('Max Wingspan (cm)')

# Highlight Snow Geese
plt.scatter(snow_geese['MaxBodyMass'], snow_geese['MaxWingspan'], color='red', s=100, label='Snow Goose')
plt.legend()
plt.show()
```

# Population Distribution
Create distribution plots and box plots to analyze variations in Snow Goose measurements and compare to population-level statistics.

```python
# Population Distribution

# Distribution plot for Snow Goose Max Length
plt.figure(figsize=(12, 6))
sns.histplot(snow_geese['MaxLength'], kde=True, color='blue', bins=10)
plt.title('Distribution of Snow Goose Max Length')
plt.xlabel('Max Length (cm)')
plt.ylabel('Frequency')
plt.show()

# Distribution plot for Snow Goose Max Wingspan
plt.figure(figsize=(12, 6))
sns.histplot(snow_geese['MaxWingspan'], kde=True, color='green', bins=10)
plt.title('Distribution of Snow Goose Max Wingspan')
plt.xlabel('Max Wingspan (cm)')
plt.ylabel('Frequency')
plt.show()

# Distribution plot for Snow Goose Max Body Mass
plt.figure(figsize=(12, 6))
sns.histplot(snow_geese['MaxBodyMass'], kde=True, color='purple', bins=10)
plt.title('Distribution of Snow Goose Max Body Mass')
plt.xlabel('Max Body Mass (g)')
plt.ylabel('Frequency')
plt.show()

# Box plot for Snow Goose Max Length
plt.figure(figsize=(12, 6))
sns.boxplot(x='Name', y='MaxLength', data=geese_species, palette='viridis')
plt.title('Box Plot of Max Length for Geese Species')
plt.ylabel('Max Length (cm)')
plt.xlabel('Geese Species')
plt.xticks(rotation=45)
plt.show()

# Box plot for Snow Goose Max Wingspan
plt.figure(figsize=(12, 6))
sns.boxplot(x='Name', y='MaxWingspan', data=geese_species, palette='viridis')
plt.title('Box Plot of Max Wingspan for Geese Species')
plt.ylabel('Max Wingspan (cm)')
plt.xlabel('Geese Species')
plt.xticks(rotation=45)
plt.show()

# Box plot for Snow Goose Max Body Mass
plt.figure(figsize=(12, 6))
sns.boxplot(x='Name', y='MaxBodyMass', data=geese_species, palette='viridis')
plt.title('Box Plot of Max Body Mass for Geese Species')
plt.ylabel('Max Body Mass (g)')
plt.xlabel('Geese Species')
plt.xticks(rotation=45)
plt.show()
```
